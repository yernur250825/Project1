```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
CSV='phones.csv'
HOST = 'https://www.sulpak.kz/'
URL = 'https://www.sulpak.kz/f/smartfoniy'
HEADERS = {
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,/;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'user-agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 13_2_3 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/13.0.3 Mobile/15E148 Safari/604.1'
}
def get_content(html):
    soup = BeautifulSoup(html, 'html.parser')
    items= soup.find_all('li', class_='tile-container')
    gadgets=[]
    for item in items:
        gadgets.append(
            {
                'title': item.find('div', class_='product-container-right-side').get_text(strip= True),
                'link_product': HOST + item.find('div', class_='product-container-right-side').find('a').get('href'),
                'price': item.find('div', class_='price-block').get_text(strip=True),
                'img':item.find('img', class_='src')
            }
        )
    return gadgets
def parser():
    html = requests.get(URL, headers=HEADERS, params='')
    if html.status_code == 200:
        phones = []
        for page in range(1, 10):
            html = requests.get(URL, params={'page': page})
            phones.extend(get_content(html.text))
        print(phones)
        with open(CSV, 'w', encoding = 'UTF-8') as file:
            writer = csv.writer(file, delimiter=',')
            writer.writerow(['title', 'link_product', 'price', 'img'])
            for item in phones:
                writer.writerow([item['title'], item['link_product'], item['price'], item['img']])
    else:
        print('ERROR')
parser()
```

```python
import os
import csv
name = []
link = []
price = []
with open('phones.csv', mode ='r') as file:
  csvfile = csv.reader(file, delimiter = ',')

  for row in csvfile:
    name.append(row[0])
    link.append(row[1])
    price.append(row[2])
    
    print(row[0], "--> " + row[1], "--> " + row[2])

class Search:
  def __init__ (self, name, link, price):
    self.name = name
    self.link = link
    self.price = price
    

  def getPrices(self, searched_phone):
    lst_of_prices = []
    for i, my_phone in enumerate(self.price):
      if my_phone == searched_phone:
        lst_of_prices.append(self.price[i] + '-->' + self.link[i])
    return lst_of_prices


class User:
  def get_info(self, name1, surname, lst_of_prices):
    self.name1 = name1
    self.link = link
    self.surname = surname
    self.lst_of_prices = lst_of_prices

  def write_info(self):
    with open(path + 'user_info.txt', 'a+') as wfile:
      wfile.write(str(self.name1) + ' ' + str(self.surname) + ',' + str(self.lst_of_prices)) 

var = True 
while var:
  d = []
  name_surname = input("Enter your name and surname with space, please: ")
  search_of_phone = input('Enter your country for searching airports: ')
  user_info = User()
  lst_of_prices = Search(name, link, price).getPrices(search_of_phone)
  
  name_surname = name_surname.strip().split(' ')
  user_info.get_info(name_surname[0], name_surname[1],  lst_of_prices)
  user_info.write_info()

  try:
    length = len(lst_of_prices)
  except TypeError:
    length = 0
    
  if length != 0:
    print(lst_of_prices)
    stop = input("Would you like to continue the session? y/n: ")
    if stop == 'n':
      var = False
    else:
      pass  
```
